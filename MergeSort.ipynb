```python
import concurrent.futures
import random
```

 Sử dụng concurrent.futures.ThreadPoolExecutor để tạo một executor với thread pool vì multiprocessing không thể chạy trên môi trường jupiter notebook


Dev Hàm Merge_Sort cho phép chạy 2 tiến trình song song

```python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left_half = arr[:mid]
    right_half = arr[mid:]

    with concurrent.futures.ThreadPoolExecutor() as executor:
        left_sorted = executor.submit(merge_sort, left_half)
        right_sorted = executor.submit(merge_sort, right_half)

        left_sorted = left_sorted.result()
        right_sorted = right_sorted.result()

    return merge(left_sorted, right_sorted)
```

Hàm merge 

```python
def merge(left, right):
    merged = []
    left_index, right_index = 0, 0

    while left_index < len(left) and right_index < len(right):
        if left[left_index] < right[right_index]:
            merged.append(left[left_index])
            left_index += 1
        else:
            merged.append(right[right_index])
            right_index += 1

    # Gắn thêm các phần tử còn lại (nếu có) từ mảng trái hoặc mảng phải vào merged
    merged.extend(left[left_index:])
    merged.extend(right[right_index:])

    return merged
```
